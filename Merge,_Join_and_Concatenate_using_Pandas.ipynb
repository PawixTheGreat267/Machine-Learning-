```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io

df1 = pd.read_csv(io.BytesIO(uploaded['LOTR.csv']))
df2 = pd.read_csv(io.BytesIO(uploaded['LOTR 2.csv']))
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how='inner', on=['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how='outer')
```

```python
df1.merge(df2, how='left')
```

```python
df1.merge(df2, how='right')
```

```python
df1.merge(df2, how='cross')
```

```python
df4 = df1.join(df2, on='FellowshipID', how='outer', lsuffix='_Left', rsuffix = '_Right')
df4
```

```python
df5 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left',rsuffix = '_Right', how = 'outer')
df5
```

```python
pd.concat([df1,df2], join='outer', axis=1)
#putting one dataFrame on top of the other not like the merge and the join where putting the DataFrame next to each other
```
