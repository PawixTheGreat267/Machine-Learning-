```python
import pandas as pd
import io

df = pd.read_csv(io.BytesIO(uploaded['Flavors.csv']))
```

```python
df
```

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
group_by_frame.mean()
```

Notice it did not take the Like Column and Flavor because they are all 'string

```python
df.groupby('Base Flavor').mean() #imoa nang gi-combine this is also called aggregation/aggregating
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').min() #naa nay flavor and liked because min and max look at the first letter so diha kay naa may string so mubasa siya

#chocolate nga naa sa flavor ang first or minimum value for that base flavor, same as sa vanilla which is Cake Batter
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').sum()
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean', 'max', 'count', 'sum'],'Texture Rating':  ['mean', 'max', 'count', 'sum']})
#using aggregation pwede nimo ma modify ang imoang columns and pwde ka mag use ug dictionary para imo siyang makita specifically ang mga elements like sum, mean, min, max, count and etc.
```

```python
df.groupby(['Base Flavor', 'Liked']).mean()
```

```python
df.groupby(['Base Flavor', 'Liked']).agg({'Flavor Rating': ['mean', 'max', 'count', 'sum'],'Texture Rating':  ['mean', 'max', 'count', 'sum']})
```

```python
df.groupby('Base Flavor').describe()
```
